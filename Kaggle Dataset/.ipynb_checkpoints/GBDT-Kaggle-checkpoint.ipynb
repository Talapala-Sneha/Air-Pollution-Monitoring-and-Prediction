```python
from pandas import concat,DataFrame
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("city_hour.csv") 
df=df[df['City']=='Chennai']
df.drop(columns=['City','NOx','PM10','SO2','NO','AQI_Bucket','Benzene','Toluene','Xylene'],axis=1,inplace=True)
df.head(5)
```

```python
df.interpolate(limit_direction="both",inplace=True)
df.head(5)
```

```python
aqi_list=df['AQI'].to_list()
```

```python
def series_to_supervised(data, n_in, n_out):
	n_vars = 1 if type(data) is list else data.shape[1]
	df = DataFrame(data)
	cols, names = list(), list()
	# input sequence (t-n, ... t-1)
	for i in range(n_in, 0, -1):
		cols.append(df.shift(i))
		names += [('AQI(t-%d)' % (i)) for j in range(n_vars)]
	# forecast sequence (t, t+1, ... t+n)
	for i in range(1, n_out):
		cols.append(df.shift(-i))
		names += [('AQI(t+%d)' % (i)) for j in range(n_vars)]
	# put it all together
	agg = concat(cols, axis=1)
	agg.columns = names
	return agg

df1= series_to_supervised(aqi_list, 4, 6)
print(df1)
```

```python
df2 = pd.concat([df.reset_index(drop=True),df1.reset_index(drop=True)], axis=1)
print(df2)
```

```python
df2.tail(5)
```

```python
actual_values=[aqi_list[-5:]]
print(actual_values)
```

```python
dropnan=True
if dropnan:
	df2.dropna(inplace=True)
```

```python
df2.tail(5)
```

```python
X=np.array(df2[['PM2.5','NO2','NH3','CO','O3','AQI','AQI(t-4)','AQI(t-3)','AQI(t-2)','AQI(t-1)']])
y=np.array(df2[['AQI(t+1)','AQI(t+2)','AQI(t+3)','AQI(t+4)','AQI(t+5)']])

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
rows,cols=X_test.shape

from sklearn.ensemble import GradientBoostingRegressor

lin_reg=MultiOutputRegressor(GradientBoostingRegressor(n_estimators=15,max_depth=16)).fit(X_train, y_train)
y_pred=lin_reg.predict(X_test)
pred=[y_pred[rows-1]]
rmse = sqrt(mean_squared_error(pred,actual_values))
print(rmse)
```
